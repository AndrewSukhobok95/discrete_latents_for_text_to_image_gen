```python
import numpy as np
from matplotlib import pyplot as plt
from torchvision import transforms as torch_transforms
import torchvision.transforms.functional as TF

import sys
sys.path.append('./../../')

from datasets.mnist_loader import MNISTData
from utilities.md_mnist_utils import LabelsInfo
from notebooks.utils import show
```

```python
def description_lines(d):
    return '{:<3} {:<3} {:<3} {:<3}\n{:<3} {:<3} {:<3} {:<3}\n{:<3} {:<3} {:<3} {:<3}'.format(*d)
```

```python
data_source = MNISTData(
    img_type='md',
    root_path='/u/82/sukhoba1/unix/Desktop/TA-VQVAE/data/multi_descriptive_MNIST/',
    batch_size=4)
train_loader = data_source.get_train_loader()

json_path = '/m/home/home8/82/sukhoba1/data/Desktop/TA-VQVAE/data/multi_descriptive_MNIST/description/labels_info.json'
li = LabelsInfo(json_path)
```

```python
images, description_ids = next(iter(train_loader))
description = li.decode_ids(description_ids.tolist())
```

```python
rows = 2
columns = 4

fig = plt.figure(figsize=(16, 8))

fig.add_subplot(rows, columns, 1)
plt.text(0, 0.4, description_lines(description[0]), size=30)
plt.axis('off')

fig.add_subplot(rows, columns, 2)
plt.imshow(np.moveaxis(images.numpy()[0], 0, -1))
plt.axis('off')

fig.add_subplot(rows, columns, 3)
plt.text(0, 0.4, description_lines(description[1]), size=30)
plt.axis('off')

fig.add_subplot(rows, columns, 4)
plt.imshow(np.moveaxis(images.numpy()[1], 0, -1))
plt.axis('off')

fig.add_subplot(rows, columns, 5)
plt.text(0, 0.4, description_lines(description[2]), size=30)
plt.axis('off')

fig.add_subplot(rows, columns, 6)
plt.imshow(np.moveaxis(images.numpy()[2], 0, -1))
plt.axis('off')

fig.add_subplot(rows, columns, 7)
plt.text(0, 0.4, description_lines(description[3]), size=30)
plt.axis('off')

fig.add_subplot(rows, columns, 8)
plt.imshow(np.moveaxis(images.numpy()[3], 0, -1))
plt.axis('off')

plt.tight_layout()
```
